## 1- Utilisation de fonction pour derterminer la moyenne entre les intervalles de date

```python
sous_nutrition.shape
```

```python
'2012-2014'.split('-')
```

```python
def compute_year(span):
    values = span.split('-')
    values = np.array(values).astype(int)
    return int(values.mean())
```

```python
sous_nutrition['Année'] = sous_nutrition['Année'].apply(compute_year)
sous_nutrition

# j'ai utilisé une fonction pour calculer l'interval entre les années histoire de definir la moyenne des deux années
# j'ai separé l'intervalle des deux dates afin de determiner la myenne des deux dates. 2012 - 2014 qui me donne 2013
```

## 2- Quel est le nombre théorique de personnes qui pourraient être nourris en 2017 ?

```python
data_dispo_alimentaire_population = pd.merge(population[population['Année'] == 2017], dispo_alimentaire, on='pays', how='inner')

# j'ai utlisé une jointure interne de la dataframe population et dispo_alimentaire pour conserver toutes les informations
# des deux dataframes
```

```python
data_dispo_alimentaire_population
```

```python
data_temporaire = data_dispo_alimentaire_population.loc[data_dispo_alimentaire_population['Année'] == 2017, ['Disponibilité alimentaire (Kcal/personne/jour)','valeur (population en milliers d’habitants)']].copy()
```

```python
data_nbres_personnes = sum(data_temporaire['valeur (population en milliers d’habitants)']*data_temporaire['Disponibilité alimentaire (Kcal/personne/jour)'])/2500
```

```python
population_mondiale = population[population['Année'] == 2017]['valeur (population en milliers d’habitants)'].sum().copy()
```

```python
population_mondiale
```

```python
round(data_nbres_personnes/population_mondiale*100,1)

# le nombre théorique de personnes qui pourraient être nourris en 2017 est de 110%
```
